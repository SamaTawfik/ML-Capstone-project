# 1. Import Libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score , confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score
```

# 2. Load Data 

```python
df = pd.read_csv("processed_data.csv")
```

```python
df
```

# 3. Logistic Regression (Classification Task) 

## 3.1 Define Variables

```python
# X--> Features   Y-->Target
X = df.drop(columns='target', axis=1)
Y = df['target']

#----------------Splitting Data---------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=42)

#-----------------Scaling---------------------
# Initialize scaler
scaler = StandardScaler()
# Fit on training data only
X_train = scaler.fit_transform(X_train)
# Transform test data
X_test = scaler.transform(X_test)
```

## 3.2 Train Logistic Regression Model

```python
# Initialize Logistic Regression
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Fit on training data
logreg.fit(X_train, Y_train)  # Use scaled data
```

## 3.3 Evaluate Model

```python
# Predict on test set
y_pred_lr = logreg.predict(X_test)

# Evaluation 
acc = accuracy_score(Y_test, y_pred_lr)
acc_perc = acc * 100
cm = confusion_matrix(Y_test, y_pred_lr)
cr = classification_report(Y_test, y_pred_lr)


print("Logistic Regression Accuracy:", acc)
print("Accuracy In Percentage : {:.2f}%".format(acc_perc))
print("Confusion Matrix:\n", cm)
print("Classification Report:\n", cr)


with open(os.path.join(CLASSIFICATION_DIR , "Logistic_Reg.txt"), "w") as f:
    f.write("Model: Logistic Regression\n")
    f.write(f"Accuracy : {acc}\n")
    f.write(f"Accuracy in percentage : {acc_perc:.2f}%\n")
    f.write(f"Confusion Matrix :\n{cm}\n")
    f.write(f"Classification Report :\n{cr}\n")


```

## 3.4 Decision Tree

```python
# Initialize Decision Tree
dtc = DecisionTreeClassifier(random_state=42, max_depth=5)  

# Fit on training data
dtc.fit(X_train, Y_train) 

# Predict on test set
y_pred_dtc = dtc.predict(X_test) 

# Evaluation (compute values first)
acc = accuracy_score(Y_test, y_pred_dtc)
acc_perc = acc * 100
cm = confusion_matrix(Y_test, y_pred_dtc)
cr = classification_report(Y_test, y_pred_dtc)

print("Decision Tree Accuracy:", acc)
print("In Percentage : {:.2f}%".format(acc_perc))
print("Confusion Matrix:\n", cm)
print("Classification Report:\n", cr)


with open(os.path.join(CLASSIFICATION_DIR , "Decision_Tree_Results.txt"), "w") as f:
    f.write("Model: Decision Tree\n")
    f.write(f"Accuracy : {acc}\n")
    f.write(f"Accuracy in percentage : {acc_perc:.2f}%\n")
    f.write(f"Decision Tree Confusion Matrix :\n{cm}\n")
    f.write(f"Decision Tree Classification Report :\n{cr}\n")
```

## 3.5 Prediction Function

##### This Function allows us to predict heart disease for new input data

```python
# ---------------- Model Training ----------------
gbc = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

gbc.fit(X_train, Y_train)

# ---------------- Prediction ----------------
y_pred_gbc = gbc.predict(X_test)

# ---------------- Evaluation ----------------
acc = accuracy_score(Y_test, y_pred_gbc)
acc_perc = acc * 100
cr = classification_report(Y_test, y_pred_gbc)


print("Accuracy:", acc)
print("In Percentage : {:.2f}%".format(acc_perc))
print("\nClassification Report:\n", cr)

# ---------------- Save to file ----------------
with open(os.path.join(CLASSIFICATION_DIR , "Model_Results.txt"), "w") as f:
    f.write("Model: Gradient Boosting Classifier \n")
    f.write(f"Accuracy : {acc}\n")
    f.write(f"Accuracy in percentage : {acc_perc:.2f}%\n")
    f.write(f"Gradient Boosting Classifier Classification Report :\n{cr}\n")


# ---------------- Confusion Matrix ----------------
cm = confusion_matrix(Y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.savefig(os.path.join(CLASSIFICATION_DIR , "confusion_matrix.png"))
plt.show()
```
